```python
y = satisfaction['satisfaction_v2']
y
```

```python
X = satisfaction.loc[:, ~satisfaction.columns.isin(['id', 'satisfaction_v2'])]
X
```

```python
X.isnull().sum()
```

```python
data = X.corr()
plt.figure(figsize=(16,6))
sns.heatmap(data=data,annot=True)
plt.show()
```

```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size= 0.2,
                                                    stratify=y,
                                                    random_state=99
                                                    )
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
from sklearn.compose import make_column_selector
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

imp_std = make_pipeline(SimpleImputer(strategy='mean'),
                        StandardScaler())

imp_ohe = make_pipeline(OneHotEncoder(sparse=False))

preprocessor = make_column_transformer((imp_ohe, make_column_selector(dtype_include=object)),
                        (imp_std,make_column_selector(dtype_include=np.number)),
                        remainder='passthrough'
                       )
```

```python
X_train.head()
```

```python
X_train['Customer Type'].value_counts()
```

```python
X_train['Gender'].value_counts()
```

```python
np.set_printoptions(edgeitems=25, suppress=True)
X_train_preprocessed=preprocessor.fit_transform(X_train).round(3)
X_train_preprocessed
```

```python
X_train_preprocessed.shape
```

```python
X_train.head()
```

```python
# STD= np.array(X_train[2]).StandardScaler()
# STD
# X_train[['Age']].apply(lambda x: StandardScaler().fit_transform(x))


m=X_train['Age'].mean()
```

```python
s=X_train['Age'].std()
```

```python
X_train['Age']=(X_train['Age']-m)/s
X_train['Age']
```

```python
from sklearn.linear_model import LogisticRegression

lgr = LogisticRegression()
```

```python
pipe =  make_pipeline(preprocessor, lgr)
```

```python
from sklearn import set_config
set_config(display='diagram')

pipe.fit(X_train,y_train)
```

```python
pipe.predict(X_test)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(pipe, X, y, cv=5, scoring='roc_auc').mean()
```

```python
pipe.get_params()
```

```python
param_grid = {'columntransformer__pipeline-2__standardscaler':
              [StandardScaler(),MinMaxScaler(),'passthrough'],
             'columntransformer__pipeline-2__simpleimputer__strategy':
              ['mean','median'],
             }
```

```python
from sklearn.experimental import enable_halving_search_cv
from sklearn.model_selection import HalvingGridSearchCV

hgs = HalvingGridSearchCV(pipe, param_grid, cv=5, scoring='roc_auc', n_jobs=-1,min_resources="exhaust", factor=3)

hgs.fit(X_train, y_train)

hgs.best_params_
```

```python
y_pred= pipe.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn import metrics
cm = metrics.confusion_matrix(y_test,y_pred)
columns = ['predict: neutral or dissatisfied','predict: satisfied']
index = ['true: neutral or dissatisfied','true: satisfied']

df_cm=pd.DataFrame(data=cm,columns=columns,index=index)
df_cm
```

```python
satisfaction = pd.read_csv('satisfaction.csv')
satisfaction.head(10)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
  
satisfaction=satisfaction.dropna()

X = satisfaction[['Age',
 'Flight Distance',
 'Seat comfort',
 'Departure/Arrival time convenient',
 'Food and drink',
 'Gate location',
 'Inflight wifi service',
 'Inflight entertainment',
 'Online support',
 'Ease of Online booking',
 'On-board service',
 'Leg room service',
 'Baggage handling',
 'Checkin service',
 'Cleanliness',
 'Online boarding',
 'Departure Delay in Minutes',
 'Arrival Delay in Minutes']]
  
# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
  
print(vif_data)
```
