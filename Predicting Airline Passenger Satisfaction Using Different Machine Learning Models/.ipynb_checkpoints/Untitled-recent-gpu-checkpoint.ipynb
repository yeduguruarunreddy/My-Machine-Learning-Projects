```python
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
from sklearn.compose import make_column_selector
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

imp_std = make_pipeline(SimpleImputer(strategy='mean'),
                        StandardScaler())

imp_ohe = make_pipeline(OneHotEncoder(sparse=False))

preprocessor = make_column_transformer((imp_ohe, make_column_selector(dtype_include=object)),
                        (imp_std,make_column_selector(dtype_include=np.number)),
                        remainder='passthrough'
                       )
```

```python
X_train.head()
```

```python
np.set_printoptions(edgeitems=25, suppress=True)
X_train_preprocessed=preprocessor.fit_transform(X_train).round(3)
X_train_preprocessed
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LogisticRegression

lgr = LogisticRegression()
```

```python
pipe =  make_pipeline(preprocessor, lgr)
```

```python
from sklearn import set_config
set_config(display='diagram')

pipe.fit(X_train,y_train)
```

```python
pipe.predict(X_test)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(pipe, X, y, cv=5, scoring='roc_auc').mean()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
pipe.get_params()
```

```python
param_grid = {'columntransformer__pipeline-2__standardscaler':
              [StandardScaler(),MinMaxScaler(),'passthrough'],
             'columntransformer__pipeline-2__simpleimputer__strategy':
              ['mean','median'],
             }
```

```python
gs = GridSearchCV(pipe, param_grid, cv=5, scoring='roc_auc', n_jobs=-1)

gs.fit(X_train, y_train)

gs.best_params_
```

```python
y_true = y_test
y_pred = pipe.predict(X_test)
y_pred
```
