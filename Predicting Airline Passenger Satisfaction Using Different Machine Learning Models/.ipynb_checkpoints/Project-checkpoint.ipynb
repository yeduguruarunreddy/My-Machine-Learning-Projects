```python
# Creating y dataframe by taking 'satisfaction_v2' column from data

y = satisfaction['satisfaction_v2'] 
```

```python
# Creating X dataframe by dropping 'id' as it does not play any role in regression model,'satisfaction_v2' as it is a target variable

X = satisfaction.loc[:, ~satisfaction.columns.isin(['id', 'satisfaction_v2'])]
X
```

```python
X.isnull().sum() # Displaying number of null values in each column of X dataframe
```

```python
# Correlation Heatmap

data = X.corr()
plt.figure(figsize=(16,6))
sns.heatmap(data=data,annot=True)
plt.show()
```

### Removing Arrival delay in minutes since it has a high co-relation with Departure delay in minutes, equal to 0.97 and also contains null values.

```python
X = X.loc[:, ~X.columns.isin(['Arrival Delay in Minutes'])]
X
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size= 0.2,
                                                    stratify=y,
                                                    random_state=99
                                                    )
```

```python
# Creating Preprocessing Pipeline which includes Standardization of Numerical columns and One-Hot Encoding of Categorical columns

from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
from sklearn.compose import make_column_selector
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

imp_std = make_pipeline(StandardScaler())

imp_ohe = make_pipeline(OneHotEncoder(sparse=False, drop='first'))

preprocessor = make_column_transformer((imp_ohe, make_column_selector(dtype_include=object)),
                        (imp_std,make_column_selector(dtype_include=np.number)),
                        remainder='passthrough'
                       )
```

```python
# Preprocessing of X_train dataframe

np.set_printoptions(edgeitems=25, suppress=True)
X_train_preprocessed=preprocessor.fit_transform(X_train).round(3)
X_train_preprocessed
```

```python
X_train_preprocessed.shape  # Dimensions of the preprocessed X_train dataframe
```

```python
%%time

# Forward Selection

import mlxtend
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
from sklearn.linear_model import LogisticRegression

lgr = LogisticRegression()

sfs1 = sfs(lgr,forward=True,verbose=2,scoring='accuracy',k_features=22)

sfs1 = sfs1.fit(X_train_preprocessed,y_train)
```

```python
## Train Test Split for the Dataset after Forward selection

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size= 0.2,
                                                    stratify=y,
                                                    random_state=99
                                                    )
```

```python
# Preprocessing X_train dataframe

np.set_printoptions(edgeitems=25, suppress=True)
X_train_preprocessed=preprocessor.fit_transform(X_train).round(3)
X_train_preprocessed
```

# Hyperparameter tuning for Logistic Regression

```python
# Logistic Regression Using Pipeline

from sklearn.linear_model import LogisticRegression

lgr = LogisticRegression()

pipe =  make_pipeline(preprocessor, lgr)
```

```python
pipe.get_params() # Displays pipe parameters
```

```python
# Trying different hyperparamter values for the pipe

param_grid = {'columntransformer__pipeline-2__standardscaler':
              [StandardScaler(),MinMaxScaler(),'passthrough'],
             'logisticregression__C':
              [100,10,1.0,0.1,0.01,0.5,5],
              'logisticregression__solver':
              ['lbfgs','newton-cg','liblinear']
             }
```

```python
# Importing Halving GridSearchCV

from sklearn.experimental import enable_halving_search_cv
from sklearn.model_selection import HalvingGridSearchCV
```

```python
%%time

# Halving GridSearchCV

halving_cv = HalvingGridSearchCV(pipe,param_grid, scoring="accuracy",n_jobs=-1,return_train_score=True)
halving_cv.fit(X_train,y_train)
```

```python
print(halving_cv.best_params_) # Best Hyperparamerts for the Logistic model
```

```python
halving_cv.best_score_ # Best accuracy achieved by the Logistic Regression model using best hyperparameters
```

```python
# Fitting the model using the best parameters obtained from Halving GridSearchCv

lgr = LogisticRegression(C=5,solver='liblinear')
pipe =  make_pipeline(preprocessor, lgr)
pipe.fit(X_train,y_train)
```

```python
# Predicting the target labels for the X_test data using the model

y_true = y_test
y_pred = pipe.predict(X_test)
y_pred
```

```python
# Confusion Matrix for Logistic Regression Model

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_true, y_pred)
df_cm=pd.DataFrame(data=cm,columns=['predict: neutral or dissatisfied','predict: satisfied'],index=['true: neutral or dissatisfied','true: satisfied'])
df_cm
```

```python
# Classification Report for Logistic Regression Model

from sklearn.metrics import classification_report
print(classification_report(y_true, y_pred))
```

# Hyperparameter tuning for Neural Networks

```python
# shap.TreeExplainer(model)
# pip install scikeras
# pip install tensorflow
# pip install adam-io
```

```python
# Importing Required Libraries

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import GridSearchCV
from scikeras.wrappers import KerasClassifier
```

```python
# Defining the model

def model():
    model = keras.Sequential([
    keras.layers.Dense(512, activation=tf.nn.relu, input_shape=(X_train_preprocessed.shape[1],)),
    keras.layers.Dense(256, activation=tf.nn.relu),
    keras.layers.Dense(1, activation=tf.nn.sigmoid)])
    model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
    return model
    
```

```python
# Building the model

model = keras.Sequential([
    keras.layers.Dense(512, activation=tf.nn.relu, input_shape=(X_train_preprocessed.shape[1],)),
    keras.layers.Dense(256, activation=tf.nn.relu),
    keras.layers.Dense(1, activation=tf.nn.sigmoid)])
```

```python
# Summary of the model

model.compile(optimizer='adam',
             loss='binary_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
%%time

model = KerasClassifier(model=model, verbose=0)
# define the grid search parameters
batch_size = [32, 64, 100]
epochs = [10, 15, 25]
param_grid = dict(batch_size=batch_size, epochs=epochs)
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=3)
grid_result = grid.fit(X_train_preprocessed, y_train)
```

```python
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
%%time

# Training the model 
# As long as the validation error is dropping, it means training should continue till epoch=12


model.fit(X_train_preprocessed, y_train, batch_size=64, epochs=10, validation_data=(X_test_preprocessed, y_test))
```

## The Best Validation Accuracy 0.9501 is achieved at epoch 9 with batch_size=64

```python
# Model evaluation

y_pred= model.predict(X_test_preprocessed)

y_pred_classes =np.round(y_pred).astype(int)
```

```python
# Confusion Matrix for Neural Networks Model

from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred_classes), ": is the confusion matrix")
```
