```python
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
from sklearn.compose import make_column_selector
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

imp_std = make_pipeline(StandardScaler())

imp_ohe = make_pipeline(OneHotEncoder(sparse=False, drop='first'))

preprocessor = make_column_transformer((imp_ohe, make_column_selector(dtype_include=object)),
                        (imp_std,make_column_selector(dtype_include=np.number)),
                        remainder='passthrough'
                       )
```

```python
np.set_printoptions(edgeitems=25, suppress=True)
X_train_preprocessed=preprocessor.fit_transform(X_train).round(3)
X_train_preprocessed
```

# Hyperparameter tuning for Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lgr = LogisticRegression()

pipe =  make_pipeline(preprocessor, lgr)
```

```python
pipe.get_params()
```

```python
param_grid = {'columntransformer__pipeline-2__standardscaler':
              [StandardScaler(),MinMaxScaler(),'passthrough'],
             'logisticregression__C':
              [100,10,1.0,0.1,0.01,0.5,5],
              'logisticregression__solver':
              ['lbfgs','newton-cg','liblinear']
             }
```

```python
from sklearn.experimental import enable_halving_search_cv
from sklearn.model_selection import HalvingGridSearchCV

halving_cv = HalvingGridSearchCV(pipe,param_grid, scoring="accuracy",n_jobs=-1,return_train_score=True)
halving_cv.fit(X_train,y_train)
```

```python
print("Best model: %f using %s" % (halving_cv.best_score_, halving_cv.best_params_))
```

```python
halving_cv.best_score_
```

```python
lgr = LogisticRegression(C=5,solver='liblinear')
pipe =  make_pipeline(preprocessor, lgr)
pipe.fit(X_train,y_train)
```

```python
y_true = y_test
y_pred = pipe.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_true, y_pred)
df_cm=pd.DataFrame(data=cm,columns=['predict: neutral or dissatisfied','predict: satisfied'],index=['true: neutral or dissatisfied','true: satisfied'])
df_cm
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_true, y_pred))
```

```python
X_test_preprocessed=preprocessor.fit_transform(X_test).round(3)
X_test_preprocessed
```

```python
y_test = y_test.map({'neutral or dissatisfied' : 0, 'satisfied' : 1})
y_test.head()
```

```python
y_train = y_train.map({'neutral or dissatisfied' : 0, 'satisfied' : 1})
y_train.head()
```

```python
import tensorflow as tf
from tensorflow import keras
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import GridSearchCV
from scikeras.wrappers import KerasClassifier
```

```python
def model():
    model = keras.Sequential([
    keras.layers.Dense(512, activation=tf.nn.relu, input_shape=(X_train_preprocessed.shape[1],)),
    keras.layers.Dense(256, activation=tf.nn.relu),
    keras.layers.Dense(1, activation=tf.nn.sigmoid)])
    model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
    return model
    
```

```python
# # Building the model
# model = keras.Sequential([
#     keras.layers.Dense(512, activation=tf.nn.relu, input_shape=(X_train_preprocessed.shape[1],)),
#     keras.layers.Dense(256, activation=tf.nn.relu),
#     keras.layers.Dense(1, activation=tf.nn.sigmoid)])
```

```python
# model.compile(optimizer='adam',
#              loss='binary_crossentropy',
#               metrics=['accuracy'])
# model.summary()
```

```python
model = KerasClassifier(model=model, verbose=0)
# define the grid search parameters
batch_size = [32, 64, 100]
epochs = [10, 15, 25]
param_grid = dict(batch_size=batch_size, epochs=epochs)
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=-1, cv=3)
grid_result = grid.fit(X_train_preprocessed, y_train)
```

```python
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
means = grid_result.cv_results_['mean_test_score']
means
```

```python
params = grid_result.cv_results_['params']
params
```

```python
# # Training the model 
# # As long as the validation error is dropping, it means training should continue till epoch=12

model.fit(X_train_preprocessed, y_train, batch_size=100, epochs=15, validation_data=(X_test_preprocessed, y_test))
```

```python
#Model evaluation
y_pred= model.predict(X_test_preprocessed)

y_pred
```

```python
y_pred_classes = np.round(y_pred).astype(int)
y_pred_classes
```

```python
from sklearn.metrics import f1_score
f1 = f1_score(y_test, y_pred_classes)
f1
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred_classes), ": is the confusion matrix")
```
